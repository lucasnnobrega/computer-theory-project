c)  Listar todos os links para outros artigos da Wikip ́edia que s ̃ao citados no conte ́udo do artigo.

````python
html = driver.page_source
````

````python
html[:100]
````

````python
pattern = r"(https?://.*)\""
````

````python
answer = re.findall(pattern, html)
````

````python
#answer
````

````python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
````

````python
# Funciona, entretanto eh pra um caso especifico pois nao filtra pela tag
#pattern = r"(?:src=\"(.+?)\")"
#pattern = r"<img[\w\d\ \=\"\'\-]*(?:src=\"(.+?)\")"
#pattern = r"<img[\w\ \=\"\'\-]*src=\"[/\w\-\_]*/(.*?)\""
pattern = r"<img[\w\ \=\"\'\-\.]*src=\"[/\w\-\_\.\:]*/(.*?)\""
````

````python
result_golden_model = []
for link in soup.find_all('img'):
    print("  ")
    print(link)
    print("|", str(link), "|", sep="")
    print(link.get('src'))
    result_golden_model.append(re.findall(pattern, str(link)))
    print(link.get('src').split("/")[-1])
    print("result", result_golden_model[-1])
    print("  ")
````

# A e HREF

````python
pattern = r"<a[\w\ \=\"\'\-\.]*href=\"[/\w\-\_\.\:\#]*/(.*?)\""
````

````python
#pattern = r'src=\"((?:https?)?.*/.+?)\" '
#pattern = r"(?:href=\"(.+?)\")"
#pattern = r"<a[\w\d\ \=\"\'\-]*(?:href=\"(.+?)\")"
````

````python
for link in soup.find_all('a'):
    print("  ")
    #print(link)
    print("|", str(link), "|", sep="")
    print(link.get('href'))
    result = re.findall(pattern, str(link))
    print("result", result)
    print("  ")
````
